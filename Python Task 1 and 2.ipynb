```python
import pandas as pd
import numpy as np
import time
import matplotlib.pyplot as plt
```

```python
#Python Code for Tree
class Node:
    
    def __init__(self, v):
        self.v = v
        self.left = None
        self.right = None
        
def insert(root, node):
    
    if root is None:
        return node
    
    if root.v > node.v:
        root.left = insert(root.left,node)

    elif root.v < node.v:
        root.right = insert(root.right,node)

    return root
def display(root):
    if root:
        display(root.left)
        print(root.v)
        display(root.right)

val = [20,30,11,15,14,23,96]

root = Node(val[0])

for i in val:
    insert(root,Node(i))
    
display(root)
```

```python
#BFS
def BFS(goal):
    explored = []
    q = []
    q.append(root)


    while len(q)>0:

        #print(q)
        
        qq = [i.v for i in q]
        print(qq)
        
        node = q.pop(0)
        
        explored.append(node.v)
        
        print('POP: ',node.v)
        print()

        if node.v == goal:
            print('Result:  ',explored)
            break


        if node.left is not None:
            q.append(node.left)

        if node.right is not None:
            q.append(node.right)

BFS(96)
```

```python
#DFS
def DFS(goal):
    explored = []
    s = []
    s.append(root)


    while len(s)>0:

        
        
        ss = [i.v for i in s]
        print(ss)
        
        node = s.pop()
        
        explored.append(node.v)
        
        print('POP: ',node.v)
        print()

        if node.v == goal:
            print('Result:  ',explored)
            break

        if node.right is not None:
            s.append(node.right)
            
        if node.left is not None:
            s.append(node.left)



DFS(96)
```

## Graph

```python
d = {'s':[('a',3),('b',2),('c',4)],
    'a':[('d',1)],
    'b':[('e',8),('f',7),('d',1)],
    'c':[('e',1),('j',3)],
    'd':[('f',1),('h',1)],
    'e':[('g',1)],
    'f':[('g',5)],
    'h':[('g',1),('f',3)],
    'j':[('e',2)],
    'g':[]}
```

```python
def my(x):
    return x[1]
def build_dcit(lis):
    di  = {}

    for i in lis:
        name = i[0]
        cost = i[1]
        di[name] = cost
    return di

def GBFS(start,goal):
    
    q = []
    start_val = d.get(start)
    q.append((start,start_val))
    explored = []
    expanded = []
    
    while len(q)>0:


        node = q.pop(0)

        if node[0] not in explored:
            explored.append(node[0])
        
        if node[0] == goal:
            print('Result :',explored,expanded,len(expanded))
            return

        child = d.get(node[0])
 
        for i in child:
            n_key = i[0]
            n_val = d.get(n_key)
            n_tuple = n_key,n_val   

            if i[0] not in explored and i[0] not in build_dcit(q):
                q.append(n_tuple)
        expanded.append(node[0])
            
        q = sorted(q,key= my)
    return explored,expanded,len(expanded)
        
    
GBFS('s','g')
```

```python
def my(x):
    return x[1]

def build_dict(lst):
    di = {}
    for i in lst:
        name = i[0]
        cost = i[1]
        di[name] = cost
    return di

def GDFS(start, goal):
    stack = []
    start_val = d.get(start)
    stack.append((start, start_val))
    explored = []
    expanded = []
    
    while len(stack) > 0:
        node = stack.pop()

        if node[0] not in explored:
            explored.append(node[0])
        
        if node[0] == goal:
            print('Result :', explored, expanded, len(expanded))
            return

        child = d.get(node[0])

        for i in child:
            n_key = i[0]
            n_val = d.get(n_key)
            n_tuple = n_key, n_val   

            if i[0] not in explored and i[0] not in build_dict(stack):
                stack.append(n_tuple)
        expanded.append(node[0])

        stack = sorted(stack, key=my, reverse=True)
    return explored, expanded, len(expanded)
GDFS('s','g')
```
