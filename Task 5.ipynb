```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
MyGraph = nx.Graph()
MyGraph.add_node('Arad')
MyGraph.add_node('Zerind')
MyGraph.add_node('Oradea')
MyGraph.add_node('Sibiu')
MyGraph.add_node('Timisoara')
MyGraph.add_node('Lugoj')
MyGraph.add_edge('Arad','Zerind',weight = 75)
MyGraph.add_edge('Arad','Sibiu',weight = 140)
MyGraph.add_edge('Arad','Timisoara',weight = 118)
MyGraph.add_edge('Zerind','Oradea',weight = 71)
MyGraph.add_edge('Sibiu','Oradea',weight = 151)
MyGraph.add_edge('Lugoj','Timisoara',weight = 111)
```

```python
 nx.draw(MyGraph, with_labels = True,node_color = 'yellow', font_weight = 'bold', node_size = 4000, width = 2, edge_color = 'red',edgecolors = 'black')


# nx.draw_random(MyGraph, with_labels = True,node_color = 'yellow', font_weight = 'bold', node_size = 4000, width = 2, edge_color = 'red',edgecolors = 'black')
# nx.draw_circular(MyGraph, with_labels = True,node_color = 'yellow', font_weight = 'bold', node_size = 4000, width = 2, edge_color = 'red',edgecolors = 'black')
# nx.draw_spectral(MyGraph, with_labels = True,node_color = 'yellow', font_weight = 'bold', node_size = 4000, width = 2, edge_color = 'red',edgecolors = 'black')
plt.show()
```

```python

G = nx.Graph()

# Another implementation
G.add_nodes_from(['Arad', 'Zerind', 'Oradea', 'Sibiu', 'Timisoara',
                  'Lugoj', 'Mehadia', 'Drobeta', 'Craiova', 'Rimnicu', 'Pitesti',
                  'Fagaras', 'Bucharest', 'Giurgiu'])


G.add_edge('Arad', 'Zerind', weight=75)
G.add_edge('Arad', 'Sibiu', weight=140)
G.add_edge('Arad', 'Timisoara', weight=118)
G.add_edge('Zerind', 'Oradea', weight=71)
G.add_edge('Oradea', 'Sibiu', weight=151)
G.add_edge('Timisoara', 'Lugoj', weight=111)
G.add_edge('Lugoj', 'Mehadia', weight=70)
G.add_edge('Mehadia', 'Drobeta', weight=75)
G.add_edge('Drobeta', 'Craiova', weight=120)
G.add_edge('Sibiu', 'Rimnicu', weight=80)
G.add_edge('Sibiu', 'Fagaras', weight=99)
G.add_edge('Rimnicu', 'Craiova', weight=146)
G.add_edge('Rimnicu', 'Pitesti', weight=97)
G.add_edge('Craiova', 'Pitesti', weight=138)
G.add_edge('Pitesti', 'Bucharest', weight=101)
G.add_edge('Fagaras', 'Bucharest', weight=211)
G.add_edge('Bucharest', 'Giurgiu', weight=90)
```

```python
nx.draw(G, with_labels = True,node_color = 'yellow', font_weight = 'bold', node_size = 4000, width = 2, edge_color = 'red',edgecolors = 'black')
```

```python
#final layout
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True,font_size=12,node_color = 'yellow', font_weight = 'bold', node_size = 400, width = 2, edge_color = 'red',edgecolors = 'black')
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)


plt.show()
```
